Reshaping the coloumn headers so as to maintain consistency

```python
annex2.columns = annex2.columns.str.strip().str.replace(' ', '_').str.lower()
annex3.columns = annex3.columns.str.strip().str.replace(' ', '_').str.lower()
annex4.columns = annex4.columns.str.strip().str.replace(' ', '_').str.lower()
annex1.columns = annex1.columns.str.strip().str.replace(' ', '_').str.lower()
```

We have now checked all files to understand what our files contain so what is next is to merge the data together into one big data using the join function so analysis and insights drawn can be a bit more effective
We start by identifying the fact table and we join the dimension table to the fact table

```python
sales = annex2.merge(annex1, on='item_code', how='left') # adds item name + category
sales = sales.merge(annex3, on=['item_code', 'date'], how='left') # adds wholesale price
sales = sales.merge(annex4, on='item_code', how='left') # adds loss rate
```

```python
sales.to_csv('sales.csv', index=False)
#exports the just merged data and makes it easy for sharing
```

```python
sales['revenue']= sales['quantity_sold_(kilo)'] * sales['unit_selling_price_(rmb/kg)']
sales['revenue']
```

```python
sales['margin'] = sales['unit_selling_price_(rmb/kg)'] - sales['wholesale_price_(rmb/kg)']
sales['margin']
```

```python
sales['effective_margin'] = sales['margin'] * (1 - sales['loss_rate_(%)'] / 100)
sales['effective_margin']
```

## Sales Performance



```python
total_revenue = round(sales['revenue'].sum(),2)
total_revenue
#total revenue generated
```

## Product-Level Insights

```python
sales.groupby('category_name')['category_name'].count().sort_values(ascending=False)
#most sold productcategory
```

```python
sales.groupby('category_name')['revenue'].sum().sort_values(ascending=False)
#category with most revenue
```

```python
sales.groupby('item_name_x')['quantity_sold_(kilo)'].sum().sort_values(ascending=False).head()
#top 10 selling items
```

```python
#items performing below average
item_sales = sales.groupby('item_name_x')['revenue'].sum()  #revenue of all items
avg_sales = item_sales.mean() #average sales of all item
below_avg_items = item_sales[item_sales < avg_sales].sort_values(ascending=True) #items below average
print(below_avg_items)
```

```python
item_sales.sort_values(ascending=True).head(10) #another way to check for underperforming items
```

## Time Based Analysis

```python
sales['month'] = sales['date'].dt.to_period('M') #month with highest sales

monthly_sales = sales.groupby('month')['revenue'].sum()

highest_month = monthly_sales.idxmax()
```

```python
print(highest_month) #month with highest sales
```
